```python
import pandas as pd
import numpy as np
```

```python
def compare_data(a, b, c, d, fname):
    dataA = pd.read_csv(a, index_col= False)
    dataB = pd.read_csv(b, index_col= False)
    dataC = pd.read_csv(c, index_col= False)
    dataD = pd.read_csv(d, index_col= False)
    
    if "Univariate" in a:
        dataA = dataA[~dataA['Model Category'].str.contains("AE")]
        dataB = dataB[~dataB['Model Category'].str.contains("AE")]
        dataC = dataC[~dataC['Model Category'].str.contains("AE")]
        dataD = dataD[~dataD['Model Category'].str.contains("AE")]
    
    modelCatA = dataA["Model Category"] + " " + dataA["Model"]
    modelCatB = dataB["Model Category"] + " " + dataB["Model"]
    modelCatC = dataC["Model Category"] + " " + dataC["Model"]
    modelCatD = dataD["Model Category"] + " " + dataD["Model"]
    
    loc1 = ["S1" for _ in range(0,28)]
    loc2 = ["S2" for _ in range(0,28)]
    loc3 = ["S3" for _ in range(0,28)]
    loc4 = ["S4" for _ in range(0,28)]

    r2 = pd.concat([dataA.R2, dataB.R2, dataC.R2, dataD.R2], ignore_index= True)
    rmse = pd.concat([dataA.RMSE, dataB.RMSE, dataC.RMSE, dataD.RMSE], ignore_index= True)
    mae = pd.concat([dataA.MAE, dataB.MAE, dataC.MAE, dataD.MAE], ignore_index= True)
    modelCat = pd.concat([modelCatA, modelCatB, modelCatC, modelCatD], ignore_index= True)
    loc = loc1 + loc2 + loc3 + loc4
    
    comp_data = pd.DataFrame({
    "Location" : loc,
    "Model": modelCat,
    "R2": r2,
    "RMSE": rmse,
    "MAE": mae
    })
    
    comp_data['Model'] = comp_data['Model'].str.replace('General ', '')

    comp_data.to_csv(fname, index= False)
    
    return comp_data
    
```

```python
def convert_series_to_range(series):
    series_min = np.min(series)
    series_max = np.max(series)
    new_min = -1
    new_max = 1
    converted_series = [(2 * (value - series_min) / (series_max - series_min)) - 1 for value in series]
    return converted_series

conv_r2 = convert_series_to_range(multi_doS1.R2)
multi_doS1["R2"] = conv_r2
conv_r2 = convert_series_to_range(multi_doS2.R2)
multi_doS2["R2"] = conv_r2
conv_r2 = convert_series_to_range(multi_doS3.R2)
multi_doS3["R2"] = conv_r2
conv_r2 = convert_series_to_range(multi_doS4.R2)
multi_doS4["R2"] = conv_r2
```

```python
def data_genR2(mulDO, mulBOD, uniDO, uniBOD, exDO, exBOD, fname):
    data = pd.DataFrame({
        "Location": mulDO.Location,
        "Model": mulDO.Model,
        "Multivariate DO": mulDO.R2,
        "Multivariate BOD": mulBOD.R2,
        "Univariate DO": uniDO.R2,
        "Univariate BOD": uniBOD.R2,
        "Except DO": exDO.R2,
        "Except BOD": exBOD.R2
    })
    
    t = data.T
    t.columns = t.iloc[1]
    t = t.iloc[2:-2]
    t = t.reset_index(drop = True)

    data.to_csv(fname, index= False)
    t.to_csv(f"T/{fname}", index= False)
    return data, t

def data_genRMSE(mulDO, mulBOD, uniDO, uniBOD, exDO, exBOD, fname):
    data = pd.DataFrame({
        "Location": mulDO.Location,
        "Model": mulDO.Model,
        "Multivariate DO": mulDO.RMSE,
        "Multivariate BOD": mulBOD.RMSE,
        "Univariate DO": uniDO.RMSE,
        "Univariate BOD": uniBOD.RMSE,
        "Except DO": exDO.RMSE,
        "Except BOD": exBOD.RMSE
    })
    
    t = data.T
    t.columns = t.iloc[1]
    t = t.iloc[2:-2]
    t = t.reset_index(drop = True)

    data.to_csv(fname, index= False)
    t.to_csv(f"T/{fname}", index= False)
    return data, t


def data_genMAE(mulDO, mulBOD, uniDO, uniBOD, exDO, exBOD, fname):
    data = pd.DataFrame({
        "Location": mulDO.Location,
        "Model": mulDO.Model,
        "Multivariate DO": mulDO.MAE,
        "Multivariate BOD": mulBOD.MAE,
        "Univariate DO": uniDO.MAE,
        "Univariate BOD": uniBOD.MAE,
        "Except DO": exDO.MAE,
        "Except BOD": exBOD.MAE
    })
    
    t = data.T
    t.columns = t.iloc[1]
    t = t.iloc[2:-2]
    t = t.reset_index(drop = True)

    data.to_csv(fname, index= False)
    t.to_csv(f"T/{fname}", index= False)
    return data, t
```
