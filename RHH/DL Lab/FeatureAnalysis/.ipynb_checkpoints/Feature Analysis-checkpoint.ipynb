```python
import shap
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

# Load the trained model
model = tf.keras.layers.TFSMLayer("../Multivariate/DOA/epochs100/auto_gru/", call_endpoint='serving_default')

# Load dataset
data = pd.read_csv("../data/locationA.csv", index_col="Date")
data.drop(columns="Unnamed: 0", inplace=True)

# Define features and target
feature_names = ['TempA', 'pHA', 'ECA', 'BODA', 'TNA']
target_name = 'DOA'

X = data[feature_names].astype(np.float32)
y = data[target_name].astype(np.float32)

# Set GRU input parameters
time_steps = 25
num_features = X.shape[1]
num_samples = X.shape[0]

# Ensure dataset length is divisible by time_steps
adjusted_length = (num_samples // time_steps) * time_steps
if adjusted_length == 0 or adjusted_length < time_steps:
    adjusted_length = time_steps

# Pad or truncate data
X_adjusted = np.zeros((adjusted_length, num_features))
X_adjusted[:min(num_samples, adjusted_length), :] = X.values[:adjusted_length, :]

# Reshape data to (batch_size, time_steps, features)
X_reshaped = X_adjusted.reshape(adjusted_length // time_steps, time_steps, num_features)

# Model prediction function
def model_predict(X_input):
    X_tensor = tf.convert_to_tensor(X_input, dtype=tf.float32)
    output_dict = model(X_tensor)
    if isinstance(output_dict, dict):
        return list(output_dict.values())[0].numpy()
    return output_dict.numpy()

# Select a subset of test samples
x_train_sample = X_reshaped[:50]
x_test_sample = X_reshaped[:10]

# Flatten input for SHAP KernelExplainer
x_train_flattened = x_train_sample.reshape(x_train_sample.shape[0], -1)
x_test_flattened = x_test_sample.reshape(x_test_sample.shape[0], -1)

# SHAP KernelExplainer
explainer = shap.KernelExplainer(lambda x: model_predict(x.reshape(x.shape[0], time_steps, num_features)), x_train_flattened)

# Compute SHAP values
shap_values = explainer.shap_values(x_test_flattened)

# Convert SHAP values back to 2D
shap_values = np.array(shap_values).squeeze()

# Ensure correct alignment
if shap_values.shape[1] != len(feature_names) * time_steps:
    shap_values = shap_values.reshape(-1, len(feature_names) * time_steps)

# Summary plot for feature importance
shap.summary_plot(shap_values, x_test_flattened, feature_names=[f"{f}_t{t}" for f in feature_names for t in range(time_steps)])

# Individual SHAP plots for each feature
for i, feature in enumerate(feature_names):
    shap.dependence_plot(i, shap_values, x_test_flattened, feature_names=[f"{f}_t{t}" for f in feature_names for t in range(time_steps)])
```

```python
import lime
import lime.lime_tabular
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

# Load the trained model
model = tf.keras.layers.TFSMLayer("../Multivariate/DOA/epochs100/auto_gru/", call_endpoint='serving_default')

# Load dataset
data = pd.read_csv("../data/locationA.csv", index_col="Date")
data.drop(columns="Unnamed: 0", inplace=True)

# Define feature and target columns
feature_names = ['TempA', 'pHA', 'ECA', 'BODA', 'TNA']
target_name = 'DOA'

X = data[feature_names].astype(np.float32)
y = data[target_name].astype(np.float32)

# Set time series parameters
time_steps = 25
num_features = X.shape[1]
num_samples = X.shape[0]

# Adjust dataset length to match time_steps
adjusted_length = (num_samples // time_steps) * time_steps
if adjusted_length == 0 or adjusted_length < time_steps:
    adjusted_length = time_steps

# Ensure X has a valid shape for GRU
X_adjusted = np.zeros((adjusted_length, num_features))
X_adjusted[:min(num_samples, adjusted_length), :] = X.values[:adjusted_length, :]

# Reshape data to match GRU input (batch_size, time_steps, features)
X_reshaped = X_adjusted.reshape(adjusted_length // time_steps, time_steps, num_features)

# Function to predict using the model while ensuring correct input shape
def model_predict(X_input):
    X_input = X_input.reshape(-1, time_steps, num_features)  # Reshape for GRU model
    X_tensor = tf.convert_to_tensor(X_input, dtype=tf.float32)
    output_dict = model(X_tensor)
    if isinstance(output_dict, dict):
        return list(output_dict.values())[0].numpy()
    return output_dict.numpy()

# Flatten the dataset for LIME
X_flattened = X_reshaped.reshape(X_reshaped.shape[0], -1)

# Initialize LIME Explainer
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_flattened,
    feature_names=[f"{f}_t{t}" for f in feature_names for t in range(time_steps)],
    class_names=["DOA"],
    mode="regression"
)

# Select a sample to explain
sample_index = 5  # Select any index from dataset
sample = X_flattened[sample_index]

# Explain the instance using LIME
exp = explainer.explain_instance(sample, model_predict, num_features=10)

# Visualize LIME Explanation
exp.show_in_notebook()
exp.as_pyplot_figure()
plt.show()
```

```python
import lime
import lime.lime_tabular
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

# Load the model using TFSMLayer
model = tf.keras.layers.TFSMLayer("../Multivariate/DOA/epochs100/auto_gru/", call_endpoint='serving_default')

# Load dataset
data = pd.read_csv("../data/locationA.csv", index_col="Date")
data.drop(columns="Unnamed: 0", inplace=True)

# Define feature and target columns
feature_names = ['TempA', 'pHA', 'ECA', 'BODA', 'TNA']
target_name = 'DOA'

X = data[feature_names].astype(np.float32)
y = data[target_name].astype(np.float32)

# Set time series parameters
time_steps = 25
num_features = X.shape[1]
num_samples = X.shape[0]

# Adjust dataset length to match time_steps
adjusted_length = (num_samples // time_steps) * time_steps
if adjusted_length == 0 or adjusted_length < time_steps:
    adjusted_length = time_steps

# Ensure X has a valid shape for GRU
X_adjusted = np.zeros((adjusted_length, num_features))
X_adjusted[:min(num_samples, adjusted_length), :] = X.values[:adjusted_length, :]

# Reshape data to match GRU input (batch_size, time_steps, features)
X_reshaped = X_adjusted.reshape(adjusted_length // time_steps, time_steps, num_features)

# Function to predict using the model while ensuring correct input shape
def model_predict(X_input):
    # Reshape input to (batch_size, time_steps, 1) if the model expects it
    X_input = X_input.reshape(-1, time_steps, 1)  # Adjust this based on the model's input signature
    X_tensor = tf.convert_to_tensor(X_input, dtype=tf.float32)
    output_dict = model(X_tensor)
    if isinstance(output_dict, dict):
        return list(output_dict.values())[0].numpy()
    return output_dict.numpy()

# Flatten the dataset for LIME
X_flattened = X_reshaped.reshape(X_reshaped.shape[0], -1)

# Initialize LIME Explainer
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_flattened,
    feature_names=[f"{f}_t{t}" for f in feature_names for t in range(time_steps)],
    class_names=["DOA"],
    mode="regression"
)

# Select a sample to explain
sample_index = 5
sample = X_flattened[sample_index]

# Explain the instance using LIME
exp = explainer.explain_instance(sample, model_predict, num_features=10)

# Visualize LIME Explanation
exp.show_in_notebook()
exp.as_pyplot_figure()
plt.show()
```
