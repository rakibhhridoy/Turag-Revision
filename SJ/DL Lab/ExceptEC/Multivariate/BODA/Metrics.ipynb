```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_metrics(epochs):
    fname = f"epochs{epochs}/metrics.csv"
    metrics = pd.read_csv(fname)
    return metrics
```

```python
epochs = 100
```

```python
ep100 = load_metrics(epochs)
ep100
```

```python
ep100.plot()
```

```python
sns.barplot(ep100, x= "ModelCat", y="MSE", hue= "Model", palette= "Dark2")
```
